```python
def relative_humidity_colormap():
    relative_humidity_colormap = matplotlib.colors.LinearSegmentedColormap.from_list("relative humidity", ["saddlebrown", "darkorange", "gold", "lightgoldenrodyellow", "yellowgreen", "lawngreen", "springgreen", "lime"])
    
    return relative_humidity_colormap
```

```python

init_hr = 6
now = datetime.now(UTC)
init_date = datetime(now.year, now.month, now.day, 6)
levels = np.arange(0, 101, 1)
ticks = levels[::5]
try:
    os.mkdir(f"Test Graphics")
except Exception as e:
    pass

try:
    for file in os.listdir(f"Test Graphics"):
        os.remove(f"Test Graphics/{file}")
except Exception as e:
    pass

for i in range(0, len(ds['step']), 1):
    if i == 0:
        fhr = init_date
    else:
        fhr = fhr + timedelta(hours=3)
    fname = f"Image_{i}.png"
    fig = plt.figure(figsize=(12, 12))
    fig.set_facecolor('aliceblue')
    ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
    ax.set_extent([-125, -60, 20, 50], ccrs.PlateCarree())
    ax.add_feature(cfeature.COASTLINE.with_scale('50m'), linewidth=0.75, zorder=9)
    ax.add_feature(cfeature.LAND, color='beige', zorder=1)
    ax.add_feature(cfeature.OCEAN, color='lightcyan', zorder=1)
    ax.add_feature(cfeature.BORDERS, linewidth=1, zorder=1)
    ax.add_feature(cfeature.LAKES, color='lightcyan', zorder=1)

    plt.title(f"RH [%]", fontsize=9, fontweight='bold', loc='left')
    plt.title(f"Forecast Valid: {fhr.strftime('%a %d/%H UTC')} FHR: {int(ds['step'][i].values)}", fontsize=7, fontweight='bold', loc='right')

    cs = ax.contourf(ds['longitude'], ds['latitude'], ds['relative_humidity_2m'][i, :, :], levels=levels, alpha=0.5, cmap=relative_humidity_colormap(), transform=ccrs.PlateCarree())
    cbar = fig.colorbar(cs, shrink=0.4, pad=0.01, location='right', ticks=ticks)
    fig.savefig(f"Test Graphics/{fname}", bbox_inches='tight')
    plt.close()
```

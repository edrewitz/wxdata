```python
import xarray as xr
import glob
import sys
import logging
import numpy as np

from keys import get_var_keys
from paths import get_branch_path
from conventions import standardize_var_keys
from coords import shift_longitude

sys.tracebacklimit = 0
logging.disable()
```

```python
paths = get_branch_path('GEFS0P50', 'members', 3)
```

```python
western_bound = -120
eastern_bound = -65
northern_bound = 50
southern_bound = 20

ds_list = []
for path in paths:     
    file_pattern = f"{path}/*.grib2"
    ds = xr.open_mfdataset(file_pattern, concat_dim='step', combine='nested', coords='minimal', engine='cfgrib', compat='override', decode_timedelta=False, filter_by_keys={'typeOfLevel': 'surface'})
    if np.nanmax(ds['longitude']) > 180:         
        ds = shift_longitude(ds)
    else:
        pass
    
    ds1 = ds.sel(longitude=slice(western_bound, eastern_bound, 1), latitude=slice(northern_bound, southern_bound, 1))
    
    ds_list.append(ds1)    

ds1 = xr.concat(ds_list, dim='number')

ds_list = []
for path in paths:     
    file_pattern = f"{path}/*.grib2"
    ds = xr.open_mfdataset(file_pattern, concat_dim='step', combine='nested', coords='minimal', engine='cfgrib', compat='override', decode_timedelta=False, filter_by_keys={'typeOfLevel': 'meanSea'})
    if np.nanmax(ds['longitude']) > 180:         
        ds = shift_longitude(ds)
    else:
        pass
    
    ds2 = ds.sel(longitude=slice(western_bound, eastern_bound, 1), latitude=slice(northern_bound, southern_bound, 1))
    
    ds_list.append(ds2)    

ds2 = xr.concat(ds_list, dim='number')
```
